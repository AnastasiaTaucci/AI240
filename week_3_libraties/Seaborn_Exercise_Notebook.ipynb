# Exercise 6 - Draw a Swarm Plot

A swarm plot is another way of plotting the distribution of an attribute or the joint distribution of a couple of attributes.

Create a Swarm plot of Petal length across the species.
 
sns.swarmplot(y='petal_length', x='species', data=iris, palette=species_colors)

```python
species_colors = ['#78C850','#F08030', '#6890F0']
sns.swarmplot(y='petal_length', x='species', hue='species', data=iris, palette=species_colors, size=4)
plt.show()
```

# Exercise 8 - Draw a Heatmap Plot

corrmat = iris.corr()

sns.heatmap(corrmat, annot = True, square = True)


```python
# Create correlation matrix
# .corr() only works with numbers. Before creating the correlation matrix, drop the non-numeric column:
corrmat = iris.drop("species", axis=1).corr()

# Draw heatmap
sns.heatmap(corrmat, annot = True, square = True, linewidths=0.5)
# sns.heatmap(corrmat, annot = True, cmap="coolwarm", linewidths=0.5)
plt.title('Correlation heatmap')
plt.show()
```
